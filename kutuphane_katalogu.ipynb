```
from google.colab import drive
drive.mount("/gdrive",force_remount=True)
dosya = "/gdrive/My Drive/Colab Notebooks/BBY162-2020/veritabani.txt"

KitapListesi = []
class Katalogum:
    def _init_(self):
        f = open(dosya,"r",encoding = "cp1254")
        for line in f.read().strip().split("\n"):
            kitap = line.strip().split("|")
            KitapListesi += [kitap]
        f.close()
        self.KitapListesi = KitapListesi
        KitapListesi = None

    def Listele(self):
      f = open(dosya,"r",encoding = "cp1254")
      if len(self.KitapListesi) < 1:
          print("Kayıt bulunamadı.")
          return None

      output = ""
      for kitap in self.KitapListesi:
          output += f'Eser adı: {kitap[0]}\n'
          output += f'Yazar adı: {kitap[1]}\n'
          output += f'Yayın Yılı: {kitap[2]}\n'
          output += f'ISBN numarası: {kitap[3]}\n'
          output += f'Yayı evi: {kitap[4]}\n'
          output += ("-"*60 + "\n")
      print(output) 
      f.close()

    def KitapEkle(self):
        KitapAdı = input("Eser adını giriniz: ")
        YazarAdı = input("Yazarını giriniz: ")
        yayınYılı = input("Yayın yılını giriniz: ")
        ISBN = input("ISBN no giriniz: ")
        YayınEvi = input("Yayınevini giriniz: ")
        GelenKayit= f'{YazarAdı}|{KitapAdı}|{yayınYılı}|{ISBN}|{YayınEvi}'
        f = open(dosya,"a",encoding = "cp1254")
        f.write(f'\n{GelenKayit}')
        f.close()
        self.KitapListesi.append([KitapAdı,YazarAdı,yayınYılı,ISBN,YayınEvi])

    def KitapAra(self):
        anahtarSozcuk = input("Arama yapacağınız anahtar sözcüğü giriniz :")#
        output = ""

        for kitap in self.KitapListesi:
            if len(self.KitapListesi) > 0 and anahtarSozcuk in kitap:
                output += f'Eser adı: {kitap[1]}\n'
                output += f'Yazar adı: {kitap[0]}\n'
                output += f'Yayın Yılı: {kitap[2]}\n'
                output += f'ISBN numarası: {kitap[3]}\n'
                output += f'Yayınevi: {kitap[4]}\n'
                print(output)
        if output == "":
            print("Aradığınız kitap bulunamadı.")


    def KitapSil(self):
        anahtarSozcuk = input("Silinecek kayıt hakkında anahtar sözcük yazınız: ")
        if len(self.KitapListesi) < 1:
            return None

        for kitap in self.KitapListesi:
            if anahtarSozcuk in kitap:
                self.KitapListesi.remove(kitap)
                break

        f = open(dosya,"a",encoding = "cp1254")
        f.truncate(0)
        for kitap in self.KitapListesi:
            f.write(f'{kitap[1]}|{kitap[0]}|{kitap[2]}|{kitap[3]}|{kitap[4]}')
            if not kitap == self.KitapListesi[-1]:
                f.write("\n")
        f.close()       

    
    def islemmenu(self):
        while True:
            islem = input("Eserleri listelemek için [1], eser kayıt etmek için [2], eser aramak için [3], eser silmek için [4], çıkış yapmak için [5] tuşuna basınız.\n") 
            
            if islem == "1":
                self.Listele()
            elif islem == "2":
                self.KitapEkle()
            elif islem == "3":
                self.KitapAra()
            elif islem == "4":
                self.KitapSil()
            elif islem == "5":
                print("Bizi tercih ettiğiniz için teşekkürler")
                break
            else:
                print("Yanlış bir tuşa bastınız!")

    Katalogum().islemmenu()

```
